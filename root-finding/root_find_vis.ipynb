#### Introduction
This notebook demonstrates comparing the convergence of Newton's method and the bisection method for finding the root of a function. The function used in this demonstration is $f(x) = x \sin(3x) âˆ’ e^x$. 


```python
# Problem spec:
# f(x) = x sin(3x) âˆ’ exp(x)
# Use Newton's  method with initial guess x(1) = âˆ’1.6 to converge to the solution to 10âˆ’6
# Use bisection with initial end points x = âˆ’0.7 and x = âˆ’0.4 to achieve an accuracy of 10âˆ’6

import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x*np.sin(3*x) - np.exp(x)

# Define the derivative of the function
def df(x):
    return np.sin(3*x) + 3*x*np.cos(3*x) - np.exp(x)

# Define the Newton-Raphson method
def newton(f, df, x0, tol):
    x = x0
    x_vals = [x0]
    i = 0
    f_val = f(x)
    while abs(f_val) > tol:
        f_val = f(x)
        x = x - f(x)/df(x)
        x_vals.append(x)
        i += 1
    return x_vals, i

# Define the bisection method
def bisection(f, a, b, tol):
    xmid = (a + b)/2
    x_vals = [xmid]
    i = 0
    while abs(f(xmid)) > tol:
        if f(a)*f(xmid) < 0:
            b = xmid
        else:
            a = xmid
        xmid = (a + b)/2
        x_vals.append(xmid)
        i += 1
    return x_vals, i

# Define the initial guesses
x0 = -1.6
a = -0.7
b = -0.4

# Solve for the x-value near x â‰ˆ âˆ’0.5 that satisfies f(x) = 0
x_newton, i_newton = newton(f, df, x0, 1e-6)

x_bisection, i_bisection = bisection(f, a, b, 1e-6)

# print number of iterations for Newton-Raphson and bisection to achieve 10^-6 accuracy
print('Newton-Raphson:', i_newton)
print('Bisection:', i_bisection+1)

# print the roots
print('Roots:')
print('Newton-Raphson:', x_newton[-1])
print('Bisection:', x_bisection[-1])

print('newton', 
      x_newton, '\n bisection',
      x_bisection)

print(type(x_newton), type(x_newton[0]))

# Save the results
A1 = x_newton 
A2 = x_bisection
A3= [i_newton, i_bisection+1]

print(len(A1))


# Plot the results
x = np.linspace(-2, 0, 100)
plt.scatter(x, f(x), label='f(x)', alpha=0.5, s=1)
plt.scatter(x_newton, f(np.array(x_newton)),  marker='x',
         c=np.arange(len(x_newton)),
         label='Newton-Raphson')
plt.scatter(x_bisection, f(np.array(x_bisection)),
        c= np.arange(len(x_bisection)),
          label='Bisection')
plt.xlabel('x')
plt.ylabel('f(x)')
# make the x scale log
# plt.xscale('symlog')
# add a grid
plt.grid(alpha=0.5)
plt.title('Newton-Raphson and Bisection Root Finding Results')
plt.legend(loc= 'lower left')
plt.show()

A = np.array([[1, 2], [-1, 1]])
B = np.array([[2, 0], [0, 2]])
C = np.array([[2,0,-3], [0,0,-1]])
D = np.array([[1,2], [2,3], [-1,0]])

x = np.array([[1], [0]])
y = np.array([[0], [1]])
z = np.array([[1], [2], [-1]])
```
